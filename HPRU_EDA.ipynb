```python
# Create a dictionary using 'ID' as the key
data_dict = {}

for index, row in df.iterrows():
    data_dict[row['ID']] = {
        'Year': row['Year'],
        'Country': row['Country'],
        'Host': row['Host'],
        'Specimen': row['Specimen'],
        'MLST':row['MLST']
    }

# Print the resulting dictionary
print(data_dict)
```

```python
# Iterate through all 'ID's and extract 'MLST' values
for key, value in data_dict.items():
    mlst_value = value.get('MLST', None)
    print(f"ID: {key}, MLST: {mlst_value}")
```

```python
# Iterate through all 'ID's and extract 'MLST' values
for key, value in data_dict.items():
    year_value = value.get('Year', None)
    mlst_value = value.get('MLST', None)
    print(f"ID: {key}, Year: {year_value}, MLST: {mlst_value}")
```
